```python
import tkinter as tk
from tkinter import messagebox
import speech_recognition as sr

# Initialize recognizer class
recognizer = sr.Recognizer()

# Global flag to control the recording
is_recording = False

# Function to start recording
def start_recording():
    global is_recording
    is_recording = True
    messagebox.showinfo("Recording", "Recording started. Press 'Stop' when done.")
    
    try:
        with sr.Microphone() as source:
            recognizer.adjust_for_ambient_noise(source)
            print("Recording...")
            audio = recognizer.listen(source, timeout=10)  # Adjust timeout if needed

            if not is_recording:
                return  # Stop if the recording was interrupted

            print("Transcribing...")
            text = recognizer.recognize_google(audio)
            print(f"You said: {text}")
            messagebox.showinfo("Transcription", f"You said: {text}")

    except sr.UnknownValueError:
        messagebox.showerror("Error", "Sorry, could not understand the audio.")
    except sr.RequestError as e:
        messagebox.showerror("Error", f"Could not request results; {e}")
    except Exception as e:
        messagebox.showerror("Error", f"An error occurred: {e}")

# Function to stop recording
def stop_recording():
    global is_recording
    is_recording = False
    messagebox.showinfo("Stopped", "Recording stopped.")

# Create the GUI
root = tk.Tk()
root.title("Speech Transcription")

# Create and place buttons
start_button = tk.Button(root, text="Start Recording", command=start_recording, width=20, height=2)
start_button.pack(pady=10)

stop_button = tk.Button(root, text="Stop Recording", command=stop_recording, width=20, height=2)
stop_button.pack(pady=10)

# Run the Tkinter event loop
root.mainloop()
```

```python
import speech_recognition as sr
import tkinter as tk
from tkinter import messagebox

# Initialize the recognizer
recognizer = sr.Recognizer()

# DSM-5 Depression Questionnaire (PHQ-9)
questions = [
    "Little interest or pleasure in doing things?",
    "Feeling down, depressed, or hopeless?",
    "Trouble falling or staying asleep, or sleeping too much?",
    "Feeling tired or having little energy?",
    "Poor appetite or overeating?",
    "Feeling bad about yourself, or that you are a failure, or have let yourself or your family down?",
    "Trouble concentrating on things, such as reading the newspaper or watching television?",
    "Moving or speaking so slowly that other people could have noticed? Or the opposite: being so fidgety or restless that you have been moving around a lot more than usual?",
    "Thoughts that you would be better off dead, or of hurting yourself in some way?"
]

# Mapping spoken answers to scores
answer_mapping = {
    "not at all": 0,
    "several days": 1,
    "more than half the days": 2,
    "nearly every day": 3
}

# To store the scores
scores = []

# Function to capture verbal response
def ask_question(question):
    print(question)
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        print("Please answer: 'Not at all', 'Several days', 'More than half the days', or 'Nearly every day'")
        try:
            audio = recognizer.listen(source, timeout=10)  # Adjust timeout as needed
            response = recognizer.recognize_google(audio).lower()
            print(f"You said: {response}")

            # Map spoken response to a score
            if response in answer_mapping:
                return answer_mapping[response]
            else:
                print("Sorry, I didn't understand. Please try again.")
                return ask_question(question)  # Re-ask the question if the response is invalid
        except sr.UnknownValueError:
            print("Sorry, I couldn't understand the audio.")
            return ask_question(question)  # Re-ask on error
        except sr.RequestError as e:
            print(f"Could not request results; {e}")
            return 0

# Function to conduct the questionnaire and calculate the score
def conduct_questionnaire():
    for idx, question in enumerate(questions):
        print(f"Question {idx + 1}:")
        score = ask_question(question)
        scores.append(score)
        print(f"Score for question {idx + 1}: {score}\n")
    
    # Calculate total score
    total_score = sum(scores)
    print(f"Total Score: {total_score}")
    
    # Provide feedback based on the total score
    if total_score >= 20:
        feedback = "Severe depression"
    elif 15 <= total_score < 20:
        feedback = "Moderately severe depression"
    elif 10 <= total_score < 15:
        feedback = "Moderate depression"
    elif 5 <= total_score < 10:
        feedback = "Mild depression"
    else:
        feedback = "Minimal or no depression"

    messagebox.showinfo("Result", f"Your total score is: {total_score}\nFeedback: {feedback}")

# Create a simple GUI for starting the questionnaire
def start_questionnaire():
    messagebox.showinfo("Start", "Starting the DSM-5 Depression Questionnaire")
    conduct_questionnaire()

# Create GUI using Tkinter
root = tk.Tk()
root.title("DSM-5 Depression Questionnaire")

start_button = tk.Button(root, text="Start Questionnaire", command=start_questionnaire, width=30, height=2)
start_button.pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import speech_recognition as sr
import pyttsx3
import threading

# Initialize recognizer class
recognizer = sr.Recognizer()
# Initialize TTS engine
tts_engine = pyttsx3.init()

# Global flag to control the recording
is_recording = False

# Function to speak the question
def speak_question():
    question = "How are you feeling?"
    tts_engine.say(question)
    tts_engine.runAndWait()

# Function to handle the recording process
def record_audio():
    global is_recording
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        print("Recording...")
        while is_recording:  # Continuously listen while the flag is True
            try:
                # Listen for the response
                audio = recognizer.listen(source, phrase_time_limit=5)

                # Transcribe the recognized phrase
                print("Transcribing...")
                text = recognizer.recognize_google(audio)
                print(f"You said: {text}")
                messagebox.showinfo("Transcription", f"You said: {text}")

            except sr.UnknownValueError:
                messagebox.showerror("Error", "Sorry, could not understand the audio.")
            except sr.RequestError as e:
                messagebox.showerror("Error", f"Could not request results; {e}")
            except Exception as e:
                messagebox.showerror("Error", f"An error occurred: {e}")
            if not is_recording:
                break  # Exit loop when stopped

# Function to start recording
def start_recording():
    global is_recording
    is_recording = True
    messagebox.showinfo("Recording", "Recording started. Press 'Stop' when done.")

    # Speak the question and start recording
    speak_question()

    # Use threading to avoid blocking the GUI
    threading.Thread(target=record_audio).start()

# Function to stop recording
def stop_recording():
    global is_recording
    is_recording = False
    messagebox.showinfo("Stopped", "Recording stopped.")

# Create the GUI
root = tk.Tk()
root.title("Speech Transcription")

# Create and place buttons
start_button = tk.Button(root, text="Ask 'How are you feeling?'", command=start_recording, width=30, height=2)
start_button.pack(pady=10)

stop_button = tk.Button(root, text="Stop Recording", command=stop_recording, width=30, height=2)
stop_button.pack(pady=10)

# Run the Tkinter event loop
root.mainloop()
```

```python
import tkinter as tk
import pyttsx3

# Initialize the TTS engine
tts_engine = pyttsx3.init()

# List of questions
questions = [
    "How are you feeling?",
    "What did you have for breakfast?",
    "How was your day?",
    "What are your plans for tomorrow?",
    "How do you feel about your work?",
    "What is something that makes you happy?",
    "What are your thoughts on the future?",
    "Is there anything you would like to change in your life?"
]

# Function to speak a given text
def speak(text):
    tts_engine.say(text)
    tts_engine.runAndWait()

# Function to ask a specific question based on the button clicked
def ask_question(index):
    if 0 <= index < len(questions):
        speak(questions[index])

# Create the GUI
root = tk.Tk()
root.title("Question Asker")

# Create and place buttons for each question
for i in range(len(questions)):
    question_button = tk.Button(root, text=f"Ask Question {i + 1}", command=lambda i=i: ask_question(i), width=20, height=2)
    question_button.pack(pady=5)

# Run the Tkinter event loop
root.mainloop()
```
